```python
#importing the dataset
import pandas as pd
titanic_data_path = '/kaggle/input/titanic/train_and_test2.csv'
titanic_data = pd.read_csv(titanic_data_path)
titanic_data.columns
```

```python
# Explore a pattern (women)
women = titanic_data[titanic_data['Sex'] == 1]

survival_rate_women = sum(women['2urvived']) / len(women)
print(survival_rate_women)
```

```python
# Explore a pattern (men)
men = titanic_data[titanic_data['Sex'] == 0]

survival_rate_men = sum(men['2urvived']) / len(men)
print(survival_rate_men)
```

```python
from sklearn.ensemble import RandomForestClassifier 

y = titanic_data["2urvived"]

features = ["Pclass", "Sex","Parch"]
X = pd.get_dummies(titanic_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

model.fit(X, y)
predictions = model.predict(X)

output = pd.DataFrame({'PassengerId': titanic_data.Passengerid, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Assuming the 'titanic_data' DataFrame is already defined and contains the necessary data

# Extracting the target variable 'Survived'
y = titanic_data["2urvived"]

# Defining the features for the model
features = ["Pclass", "Sex", "Parch"]

# Creating dummy variables for categorical features
X = pd.get_dummies(titanic_data[features])

# Initializing and training the RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

# Defining the variables for the passenger in question
Pclass = input("What is the passengers class? 1-2-3")
Sex = input("Male of Female?")
if Sex == "Female":
    Sex = 1
elif Sex == "Male":
    Sex = 0

Parch = input("Number of parents or children aboard?")

# Creating a DataFrame containing data for the passenger in question
passenger_data = pd.DataFrame([[Pclass, Sex, Parch]], columns=features)

# Generating predictions for the passenger in question
prediction = model.predict(passenger_data)

if prediction == 1:
    print("This passenger WILL SURVIVE!")
elif prediction == 0:
    print("This passenger WILL NOT SURVIVE!")
else:
    print("ERROR")
```
